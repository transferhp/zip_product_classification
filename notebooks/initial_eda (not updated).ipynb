```python
another_df["e_cat_l0"] = another_df['e_cat_l0'].apply(lambda x: ", ".join(x)) 
another_df["e_cat_l1"] = another_df['e_cat_l1'].apply(lambda x: ", ".join(x)) 
another_df["e_cat_l2"] = another_df['e_cat_l2'].apply(lambda x: ", ".join(x)) 
```

```python
another_df["e_cat_l0"].value_counts().plot(kind='bar', figsize=(14, 5))
```

```python
another_df["e_cat_l1"].value_counts().plot(kind='bar', figsize=(14, 5))
```

```python
another_df["e_cat_l2"].value_counts().plot(kind='bar', figsize=(14, 5))
```

```python
another_df["cat0_cat1_cat_2"] = (
    another_df["e_cat_l0"] + '|'
    + another_df["e_cat_l1"] + '|'
    + another_df["e_cat_l2"]
)

another_df["cat0_cat1_cat_2"].nunique()
```
